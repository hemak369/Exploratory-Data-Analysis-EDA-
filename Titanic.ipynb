```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create FamilySize column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

fig, axs = plt.subplots(2, 3, figsize=(18, 10))  # 2 rows, 3 columns

# Chart 1: Survival Count
ax1 = sns.countplot(x='Survived', data=df, ax=axs[0, 0])
ax1.set_title("Survival Distribution")
ax1.bar_label(ax1.containers[0])

# Chart 2: Survival by Gender
ax2 = sns.countplot(x='Survived', hue='Sex', data=df, ax=axs[0, 1])
ax2.set_title("Survival by Gender")
for container in ax2.containers:
    ax2.bar_label(container)

# Chart 3: Survival by Class
ax3 = sns.countplot(x='Survived', hue='Pclass', data=df, ax=axs[0, 2])
ax3.set_title("Survival by Class")
for container in ax3.containers:
    ax3.bar_label(container)

# Chart 4: Fare Distribution
ax4 = sns.histplot(df['Fare'], bins=40, kde=True, ax=axs[1, 0])
ax4.set_title("Fare Distribution")
for p in ax4.patches:
    height = p.get_height()
    if height > 0:
        ax4.text(p.get_x() + p.get_width()/2, height, int(height), ha='center', va='bottom')

# Chart 5: Family Size Distribution
ax5 = sns.countplot(x='FamilySize', data=df, ax=axs[1, 1])
ax5.set_title("Family Size Distribution")
ax5.bar_label(ax5.containers[0])

# Chart 6: Embarked Pie Chart
df['Embarked'].value_counts().plot.pie(
    autopct='%1.1f%%', startangle=90, ax=axs[1, 2]
)
axs[1, 2].set_title("Embarkation Port Distribution")
axs[1, 2].set_ylabel('')

fig.suptitle("Titanic Data Visualization Overview", fontsize=30, color='darkred',weight='bold')
plt.tight_layout()
plt.show()
```
